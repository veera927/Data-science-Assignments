```python
# import the necessary libraries in Python: 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# Exercise 1: Prime Numbers
def is_prime(num):
    """
    This function checks if a given number is prime.
    A prime number is greater than 1 and only divisible by 1 and itself.
    """
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):  # Optimized: check divisibility till square root of num
        if num % i == 0:
            return False
    return True

# Test the function
num = int(input("Enter a number to check if it is prime: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")


# Exercise 2: Product of Random Numbers
import random

def check_product():
    """
    This function generates two random numbers and asks the user to input the product.
    It then checks if the user's answer is correct.
    """
    num1 = random.randint(1, 10)  # Random number between 1 and 10
    num2 = random.randint(1, 10)  # Random number between 1 and 10
    correct_product = num1 * num2
    print(f"Guess the product of {num1} and {num2}: ")
    user_product = int(input("Enter your answer: "))
    
    if user_product == correct_product:
        print("Correct! Well done.")
    else:
        print(f"Oops! The correct product was {correct_product}.")

check_product()


# Exercise 3: Squares of Even/Odd Numbers
def squares_of_numbers(choice='even'):
    """
    This function prints the squares of either even or odd numbers between 100 and 200.
    The choice can be 'even' or 'odd'.
    """
    if choice == 'even':
        for num in range(100, 201):
            if num % 2 == 0:
                print(f"The square of {num} is {num ** 2}")
    elif choice == 'odd':
        for num in range(100, 201):
            if num % 2 != 0:
                print(f"The square of {num} is {num ** 2}")

# Call the function for even numbers (can change to 'odd' if desired)
squares_of_numbers(choice='even')


# Exercise 4: Word Counter
def word_counter(text):
    """
    This function counts the number of occurrences of each word in the input text.
    """
    words = text.split()  # Split the text into words
    word_count = {}
    for word in words:
        word = word.lower().strip(".")  # Convert word to lowercase and remove punctuation
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count

# Test the function with a sample text
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_counts = word_counter(input_text)
print("Word counts:")
for word, count in word_counts.items():
    print(f"'{word}': {count}")


# Exercise 5: Check for Palindrome
def is_palindrome(input_string):
    """
    This function checks if the given string is a palindrome.
    It ignores spaces, punctuation, and case differences.
    """
    # Remove spaces and punctuation, and convert to lowercase
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())
    return cleaned_string == cleaned_string[::-1]

# Test the palindrome function
test_string = input("Enter a string to check if it is a palindrome: ")
if is_palindrome(test_string):
    print(f"'{test_string}' is a palindrome.")
else:
    print(f"'{test_string}' is not a palindrome.")
```
